```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import sklearn.datasets 
from sklearn.model_selection import train_test_split 
from sklearn.datasets import load_breast_cancer
```

```python
breast_cancer_dataset = load_breast_cancer()
```

```python
df =pd.DataFrame(breast_cancer_dataset.data,columns=breast_cancer_dataset.feature_names)  

df.head()
```

```python
# adding the 'target' columns to the data frame 
df['label'] = breast_cancer_dataset.target 
```

```python
X = df.drop(columns='label',axis=1)  
y = df['label']
```

```python
X
```

```python
y
```

##  Split the data into training data and testing data 

```python
x_train,x_test,y_train,y_test =train_test_split(X,y,test_size=0.2,random_state=2)
```

```python
from sklearn.preprocessing import StandardScaler 
```

```python
sc =StandardScaler() 
x_train_std =sc.fit_transform(x_train) 
x_test_std =sc.transform(x_test) 
```

## Building the Neural Network 

```python
import tensorflow as tf 

tf.random.set_seed(3) 
from tensorflow import keras 
```

```python
# setting up the layer of my neural Network 
model =keras.Sequential([
    keras.layers.Flatten(input_shape=(30,)) ,
    keras.layers.Dense(20,activation='relu') ,
    keras.layers.Dense(2,activation='sigmoid') 
])  
```

```python
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
```

```python
## training the data for Neural Network 

history =model.fit(x_train_std,y_train,validation_split=0.1,epochs=30) 
```

```python
plt.plot(history.history['accuracy']) 
plt.plot(history.history['val_accuracy']) 

plt.title('Model Accuracy') 
plt.ylabel('Accuracy') 
plt.xlabel('epoch') 

plt.legend(['training data','validation data'],loc='lower right') 

plt.tight_layout() 
plt.show()
```

```python
plt.plot(history.history['loss']) 
plt.plot(history.history['val_loss']) 

plt.title('Model loss') 
plt.ylabel('loss') 
plt.xlabel('epoch') 

plt.legend(['training loss','validation data'],loc='upper right') 

plt.tight_layout() 
plt.show()
```

```python
loss , accuracy = model.evaluate(x_test_std,y_test)  
print(accuracy)
```

```python
y_pred =model.predict(x_test_std) 
```

```python
y_pred_labels = [np.argmax(i) for i in y_pred]
```
